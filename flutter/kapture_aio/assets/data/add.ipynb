```python
import pandas as pd
from geopy.geocoders import Nominatim

# Read the CSV data
data = pd.read_csv('WeatherPhoto.csv', header=None, names=['place', 'direction', 'url'])

# Initialize the geolocator
geolocator = Nominatim(user_agent="geoapiExercises", timeout=10)

def get_lat_lon(place):
    location = geolocator.geocode(place)
    if location:
        return pd.Series((location.latitude, location.longitude))
    else:
        return pd.Series((None, None))

# Apply the function to the 'place' column
data[['latitude', 'longitude']] = data['place'].apply(get_lat_lon)

# Save the DataFrame back to CSV
data.to_csv('WeatherPhoto_updated.csv', index=False)
```
